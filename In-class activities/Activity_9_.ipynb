```python
import requests

# Define the URL of the Bugzilla API
url = "https://bugzilla.redhat.com/rest/bug?status=CLOSED&classification=Fedora&product=Fedora&resolution=WONTFIX&resolution=DEFERRED&resolution=CURRENTRELEASE&resolution=ERRATA&resolution=NEXTRELEASE&resolution=RAWHIDE&resolution=UPSTREAM&limit=10"

# Send a GET request to the API
response = requests.get(url)
data = response.json()

# Extract and display the total number of bugs found
total_bugs = len(data['bugs'])
print(f"Total number of bugs found: {total_bugs}\n")

# Extract and display information for each bug
bugs = data['bugs']
for bug in bugs:
    bug_id = bug['id']
    component = bug['component']
    summary = bug['summary']
    status = bug['status']
   
    # Display the bug IDs (without #)
    print(f"Bug ID: {bug_id}")

    # Display the component name, without any additional characters
    print(f"Component: {component}")

    # Display the bug summary
    print(f"Summary: {summary}")

    # Display the bug status
    print(f"Status: {status}")
   
    print()
```
